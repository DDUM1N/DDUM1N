## 1. 결합분포 및 상관관계 분석(주변확률밀도함수 그래프)

```python
import matplotlib.pyplot as plt

# 결합확률질량함수 데이터
joint_pmf_data = {
    (1, 1): 0.1,
    (1, 2): 0.2,
    (1, 3): 0.2,
    (2, 1): 0.3,
    (2, 2): 0.05,
    (2, 3): 0.15
}

# X와 Y의 주변확률밀도함수를 계산
f_X = {}
f_Y = {}

# X의 주변확률밀도함수 계산
for x, y in joint_pmf_data.keys():
  f_X[x] = f_X.get(x, 0) + joint_pmf_data[(x, y)]

# Y의 주변확률밀도함수 계산
for x, y in joint_pmf_data.keys():
  f_Y[y] = f_Y.get(y, 0) + joint_pmf_data[(x, y)]

# Y축의 범위를 0부터 1까지, X축의 범위를 동일하게 설정하여 그래프를 그리기
plt.figure(figsize=(12, 5))

# Y축의 범위 설정
y_min = 0
y_max = 1

# X축의 범위 설정
x_min = 0.5
x_max = 3.5

# X의 주변확률밀도함수 그래프
plt.subplot(1, 2, 1)
plt.plot(list(f_X.keys()), list(f_X.values()), 'bo') # 점
plt.vlines(list(f_X.keys()), y_min,
list(f_X.values()), colors='blue') # 수선의 발
plt.title('Marginal PMF of X')
plt.xlabel('X value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max)+1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# Y의 주변확률밀도함수 그래프
plt.subplot(1, 2, 2)
plt.plot(list(f_Y.keys()), list(f_Y.values()), 'ro') # 점
plt.vlines(list(f_Y.keys()), y_min, list(f_Y.values()),
colors='red') # 수선의 발
plt.title('Marginal PMF of Y')
plt.xlabel('Y value')
plt.ylabel('Probability')
plt.xticks(range(int(x_min), int(x_max)+1))
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

plt.tight_layout()
plt.show()
```

- 결합확률에서 주변확률을 구할 때는 나머지 변수를 모두 더해야 함
- subplot으로 X와 Y의 주변확률을 동시에 시각화할 수 있음


## 2. 결합분포 및 상관관계 분석(공분산)

```python
import numpy as np

# 국어 성적과 수학 성적 데이터
korean_scores = [75, 65, 50, 70, 80]
math_scores = [65, 70, 70, 55, 60]

# 데이터를 NumPy 배열로 변환
data = np.array([korean_scores, math_scores])

# 공분산 계산
covariance = np.cov(data, bias=True)[0, 1]

# 결과 출력
print("Covariance:", covariance)
```

- 공분산이 음수일 때 두 변수는 반대 방향으로 변함
- numpy 배열로 변환 후 np.cov()로 공분산을 계산
- 공분산은 단위의 영향을 받음


##3. 결합분포 및 상관관계 분석(상관계수)

```python
import numpy as np
import numpy as np

# 데이터
x = np.array([4, 6, -4, 10])
y = np.array([-1, 5, 3, 18])

# 상관계수 계산
correlation = np.corrcoef(x, y)[0, 1]

# 결과 출력
print("Correlation Coefficient:", correlation)
```

- np.corrcoef()로 상관계수를 계산
- 상관계수는 -1에서 1 사이의 값으로 관계 방향과 강도를 보여줌
- 공분산과 달리 단위에 영향을 받지 않음

##연습문제 18번
다음 데이터는 하루 운동시간 와 체질량 지수 에 대한 15명의 데이터가 다음 표 8.11와 같다. 파이썬 코드를 이용하여 공분산을 구하고 이를 그래프로 그려보라.image.png

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.array([3, 2, 0, 1, 2, 4, 1, 2, 1, 3, 2, 0, 3, 1, 0]) # 운동시간
y = np.array([15, 21, 25, 23, 19, 20, 23, 19, 20, 17, 21, 28, 20, 22, 26]) # 체질량 지수

cov_xy = np.average((x-np.average(x))*(y-np.average(y)))
print("cov_xy", cov_xy)

plt.scatter(x, y, marker="o", label = "xy", s = 40) # s는 마커의 크기
plt.legend()
plt.xlabel( "x", size = 14)
plt.ylabel( "y", size = 14)

plt.grid()
plt.show()
```

- 공분산을 NumPy의 average() 함수를 사용하여 계산할 수 있다.
- matplotlib을 이용하여 그래프를 그려 두 데이터의 관계를 시각화할 수 있다.

- 여기서 상관계수가 –0.778130598295546로 음수이므로, 두 변수 간에 어느 정도 강한 음의 선형 관계가 있음을 나타낸다.
- 운동시간이 증가하면 체질량지수가 감소하는 경향이 있다고 할 수 있다. 다시 말해, 운동을 더 많이 할수록 체질량지수가 낮아지는 경향이 있음을 시사한다.
